```python
# キーボードから税抜価格の入力を受け付ける
price_excluding_tax = int(input("税抜価格を入力してください："))
price_excluding_tax
```

```python
# 税抜価格から税込価格を計算する
tax = 0.1
price_including_tax = price_excluding_tax + (price_excluding_tax * tax)
price_including_tax = int(round(price_including_tax, 0))
price_including_tax
```

```python
# 税込価格によって送料がいくらかかるかを表示する
def calc_shipping_fee(price_including_tax):
  if price_including_tax >= 2000:
    print("送料は無料です")
  else:
    print("送料として350円かかります")
    price_including_tax = price_including_tax + 350
  print(f"送料込みの価格は{price_including_tax:,}円です。")

calc_shipping_fee(price_including_tax)
```
